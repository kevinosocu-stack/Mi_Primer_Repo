```python
import requests
import json
from googletrans import Translator

translator = Translator()

def obtener_palabra_aleatoria():
    """
    Obtiene una palabra aleatoria en inglés de una API.
    """
    url = "https://random-word-api.herokuapp.com/word"
    
    try:
        # Realiza la petición GET a la API
        respuesta = requests.get(url)
        # Lanza una excepción si la petición no fue exitosa (código de estado 4xx o 5xx)
        respuesta.raise_for_status() 
        
        # Analiza la respuesta JSON
        # La API devuelve una lista, por ejemplo: ['ejemplo']
        datos = respuesta.json()
        
        if datos and isinstance(datos, list):
            return datos[0] # Devuelve el primer elemento de la lista (la palabra)
        else:
            return "No se pudo obtener una palabra válida."
            
    except requests.exceptions.RequestException as e:
        return f"Error al conectar con la API: {e}"
    except json.JSONDecodeError:
        return "Error al decodificar la respuesta JSON."

# Bienvenida e instrucciones

print("""
Bienvenido a El ahorcado, tienes que adivinar la palabra que está pensando la computadora.
La computadora pensará en una palabra aleatoria y tú tendrás que adivinarla.
Si adivinas la palabra, ganas el juego.
Si te equivocas, la computadora te dirá que esa letra no forma parte de la palabra.
Solo tienes 8 vidas, así que elige con cuidado.
""")


while True:
    print("pensando en una palabra...")
    cpuI = obtener_palabra_aleatoria() # escoge la palabra
    #print(cpuI)
    
    try:
        traduccion = translator.translate(cpuI, 'es')
        #if traduccion is not None and traduccion.text:
            #print(traduccion.text)
        #else:
            #print("❌ La traducción fue vacía o el objeto devuelto no es válido.")

    except Exception as e:
        continue
        #print(f"⚠️ Error al intentar traducir: {e}")

    cpu = traduccion.text.lower()
    #print(cpu)
    if len(cpu) <= 6. and cpu is not None: #asegura que la palabra tenga máximo 6 letras
        break

caracteres = len(cpu) #cuenta los caracteres en total
#print(caracteres)

# lista de _
palabra_adivinar= [" _ "] * caracteres #crea una lista con la cantidad de letras de la palabra usando _
print("La palabra tiene", caracteres, "letras en total:", "" .join(palabra_adivinar)) 

finalizar = False
letras_intentadas = set() #conjunto de letras que el jugador ya intento
vidas = 8
letra_encontrada = False

while not finalizar:
    
    print("Palabra hasta ahora:", "".join(palabra_adivinar))
    print(f"Te quedan {vidas} vidas.")
    
    jugador = input("Escribe la letra que crees que está en la palabra:").lower()
    
    if jugador in letras_intentadas:
        print("Ya has intentado con esa letra. Intenta con otra.")
        continue
    else:
        letras_intentadas.add(jugador)
        letra_encontrada = False
        
    for i in range(caracteres):
        if cpu[i] == jugador:
            palabra_adivinar[i] = jugador
            letra_encontrada = True
            
    if letra_encontrada:
        print("Bien hecho, la letra", jugador, "está en la palabra.")
        
        if " _ " not in "".join(palabra_adivinar):
            print("¡Felicidades! Has adivinado la palabra:", cpu)
            finalizar = True
            
    else:
        vidas = vidas - 1
        print("La letra", jugador, "no está en la palabra.")
        print("Te quedan", vidas, "vidas.")
        
        if vidas == 0:
            print("Se te acabaron las vidas. La palabra era:", cpu)
            finalizar = True
# Me ayudo el doc 
```
